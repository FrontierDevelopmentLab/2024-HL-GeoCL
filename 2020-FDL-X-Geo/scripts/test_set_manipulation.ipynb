```python
from astropy.time import Time
import h5py
import os
```

## View the test set hdf files

```python
file_to_view = "data_local/weimer/2010/TimeStepGeomagnetic20100405_start_and_end.h5"
weimer = {}
with h5py.File(file_to_view, 'r') as f:
    for k in f.keys():
        weimer[k] = f.get(k)[:]
print(weimer.keys())
```

```python
print(weimer["JDTIMES"])
human_readable = [Time(wtime, format="jd").to_value("isot") for wtime in weimer["JDTIMES"]]
print(human_readable[:4], human_readable[-4:])
```

## Add a storm to the test set

```python
stime = {"year": 2010, "month": 4, "day": 5, "hour": 0, "minute": 0}
etime = {"year": 2010, "month": 4, "day": 6, "hour": 0, "minute": 0}

jd_stime = Time(stime, format='ymdhms').to_value("jd")
jd_etime = Time(etime, format='ymdhms').to_value("jd")
print(jd_stime, jd_etime)
```

```python
file_to_save = f"TimeStepGeomagnetic{Time(jd_stime, format='jd').to_value('isot')[:10].replace('-', '')}_start_and_end.h5"

assert(stime["year"] == etime["year"])
try:
    os.mkdir("data_local/weimer/"+str(stime["year"]))
except FileExistsError:
    pass

storm_times = [jd_stime, jd_etime]
with h5py.File("data_local/weimer/"+str(stime['year'])+'/'+file_to_save, "w") as f:
    dset = f.create_dataset("JDTIMES", shape=(2,))
    dset[0] = storm_times[0]
    dset[1] = storm_times[1]
```
